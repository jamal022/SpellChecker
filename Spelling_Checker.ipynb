```python
class SpellChecker:
    def __init__(self, dictionary_file):
        self.dictionary = set() # Set it suitable data structure for checking if the word exists or not 
        self.load_dictionary(dictionary_file)
        
    def load_dictionary(self, dictionary_file):
        with open(dictionary_file, "r",encoding="utf-8", errors="ignore") as file:
            for word in file:
                self.dictionary.add(word.strip())

    def find_nearest_words(self, word):
        nearest_words = []
        if word not in self.dictionary:
            words_before = [w for w in self.dictionary if w < word]
            words_after = [w for w in self.dictionary if w > word]
            nearest_words = sorted(words_before)[-2:] + sorted(words_after)[:2]
        return nearest_words
    
    def add_word_to_dictionary(self, word):
        self.dictionary.add(word)
```

```python
dictionary_file_path = "dictionary.txt"
```

```python
spell_checker = SpellChecker(dictionary_file_path)
```

Since the size of the dictionary is kinda big so lets create a function to check if the word already in the dictionary or not to check find_nearest_words logic

```python
def is_word_in_dictionary(dictionary, word):
    return 1 if word in dictionary else 0  # 1 if there word exists and 0 otherwise 
```

```python
is_word_in_dictionary(spell_checker.dictionary, "quars")
```

So as shown above the word "quars" not mentioned in our dictionary so we will use it as our test case 

```python
# Find nearest 4 words for an input word not in the dictionary
input_word = "quars"
nearest_words = spell_checker.find_nearest_words(input_word)
print(f"Nearest words to '{input_word}': {nearest_words}")
```

```python
# Add a new word to the dictionary
new_word = "quars"
spell_checker.add_word_to_dictionary(new_word)
print(f"Dictionary after adding '{new_word}': {spell_checker.dictionary}")
```

Check now after adding "quars" word to our dictionary 

```python
is_word_in_dictionary(spell_checker.dictionary, "quars")
```
